## 1. Solving Simple Cases

As my UC Berkeley Calc 2 professor Ole Hald used to say in a thick Danish accent while demonstrating concepts on the blackboard, "I'm going to use the dumbest example I can think of." So let's forget about the 100 passengers and consider an easy example with a plane that only holds 2 passengers. 

I'm going to use the notation $P(n \rightarrow m)$ to denote the probability that Passenger $n$ ends up sitting in the seat assigned to Passenger $m$, and I'll use $P(n \not \rightarrow m)$ to denote the probability that this does NOT happen, for given arbitrary passengers $n$ and $m$. We'll label each passenger by the position they have in line. So the first passenger to enter the plane (i.e. the idiot who decides to just sit anywhere!) will be labeled Passenger 1, and so on. I will sometimes refer to passengers by their numbers, for short (e.g. "$n$ sits in seat $m$".)

### 2-Seat Plane

In the case with 2 passengers, the probability that you (Passenger 2) get to sit in your assigned seat is equal to the probability that Passenger 1, upon randomly choosing a seat, picks the one that isn't yours. This probability is of course $\boxed{1/2}$.

### 3-Seat Plane

Now let's try the case where there are 3 seats on the plane. Now, the probability that you (Passenger 3) get to sit in your assigned seat is equal to:

\begin{align}
P(3 \rightarrow 3) &= P(1 \not \rightarrow 3 \text{ and } 2 \not \rightarrow 3) \\
&= 1 - P(1 \rightarrow 3 \text{ or } 2 \rightarrow 3) \\
&= 1 - \left[P(1 \rightarrow 3) + P(2 \rightarrow 3)\right]
\end{align}

where the third line holds because of the mutual exclusivity of the two events (Passengers 1 and 2 cannot both sit in seat 3).

Now, $P(1 \rightarrow 3) = 1/3$ because Passenger 1 randomly chooses one of the three seats on the plane. Meanwhile, the only way that Passenger 2 could possibly end up taking your seat is if Passenger 2's seat was taken by Passenger 1. Hence, $P(2 \rightarrow 3) = P(2 \rightarrow 3 | 1 \rightarrow 2)P(1 \rightarrow 2) = P(2 \rightarrow 3 | 1 \rightarrow 2)\cdot(1/3)$. Now, note that if Passenger 1 has taken Passenger 2's seat, Passenger 2 is left with a random choice between Passenger 1's seat and your seat, so $P(2 \rightarrow 3 | 1 \rightarrow 2) = 1/2$. So $P(2 \rightarrow 3) = (1/2)\cdot(1/3) = 1/6.$

Therefore, 

\begin{align}
P(3 \rightarrow 3) &= 1 - \left[P(1 \rightarrow 3) + P(2 \rightarrow 3)\right] \\
&= 1 - \left[\frac{1}{3} + \frac{1}{6}\right] \\
&= \boxed{\frac{1}{2}}
\end{align}

### 4-Seat Plane

In the case where the plane has four seats, and you are Passenger 4, your chance of getting your assigned seat is, similarly to the previous example:

\begin{align}
P(4 \rightarrow 4) &= 1 - P(\text{1, 2, or 3} \rightarrow 4) \\
&= 1 - \left[P(1 \rightarrow 4) + P(2 \rightarrow 4) + P(3 \rightarrow 4)\right]
\end{align}

where the part in square brackets again follows because of mutual exclusivity of all the different ways in which your seat can be taken by someone before you.

By this point we know immediately that $P(1 \rightarrow 4) = 1/4$ since Passenger 1 has a random choice of four seats. Meanwhile, once again, Passenger 2 can take seat 4 only if Passenger 1 took Passenger 2's seat, so $P(2 \rightarrow 4) = P(2 \rightarrow 4 | 1 \rightarrow 2)P(1\rightarrow 2) = P(2 \rightarrow 4 | 1 \rightarrow 2) \cdot (1/4) = (1/3)\cdot(1/4) = 1/12$

Now consider the possibility that Passenger 3 takes seat 4. This can only happen if either Passenger 1 took seat 3 or Passenger 2 took seat 3.

\begin{align}
P(3 \rightarrow 4) &= P(3 \rightarrow 4|\text{1 or 2 $\rightarrow$ 3})\cdot P(\text{1 or 2 $\rightarrow$ 3}) \\
&= \underbrace{\frac{1}{2}}_{\text{Passenger 3 has a random} \\ \text{choice of two seats}} \cdot \underbrace{\left[P(1 \rightarrow 3) + P(2 \rightarrow 3)\right]}_{\text{mutual exclusivity}} \\
&= \frac{1}{2} \cdot \left[\frac{1}{4} + \underbrace{P(2 \rightarrow 3 | 1 \rightarrow 2)\cdot P(1\rightarrow 2)}_{\text{2 can only take 3's seat if 1 took 2's seat}}\right] \\
&= \frac{1}{2} \cdot \left[\frac{1}{4} + \underbrace{\frac{1}{3}}_{\text{2 has a random} \\ \text{choice of three seats}} \cdot \frac{1}{4} \right] \\
&= \frac{1}{6}
\end{align}

Now combining all of these:

\begin{align}
P(4 \rightarrow 4) &= 1 - \left[P(1 \rightarrow 4) + P(2 \rightarrow 4) + P(3 \rightarrow 4)\right] \\
&= 1 - \left[\frac{1}{4} + \frac{1}{12} + \frac{1}{6}\right] \\
&= \boxed{\frac{1}{2}}
\end{align}

It seems that there is a pattern here. While we haven't proven it will hold for all cases, in each of these simple examples, the probability of you, the last person in line, getting your seat is 1/2. Let's try attacking the 100-seat problem now.

## 2. Solving the 100-Seat Case

There is one crucial fact to realize about this problem, regardless of how many seats are on the plane: When Passenger 1 picks a seat at random (let $n$ be the seat he picked, and ignore the trivial case in which $n=1$, i.e. where he picks his assigned seat), every subsequent passenger from 2 through $n-1$ is totally unaffected and therefore becomes irrelevant to the problem; each one of them is able to sit in their assigned seat and therefore does so. 

The first displaced passenger is $n$. Passenger $n$ enters a plane where:

* The number of seats available is equal to the number of people who have yet to be seated (Passenger $n$ and everyone after them)
* The seats assigned to everyone after Passenger $n$ have not yet been taken.
* The first passenger to enter (i.e., Passenger $n$) is guaranteed to choose a seat at random.

These are exactly the same circumstances that hold for the simple cases we solved above!  

Consider the case where Passenger 1 chooses the seat belonging to Passenger 99 (i.e. the second-to-last passenger). The probability of you (Passenger 100) getting to sit in seat 100 conditioned on this event is $1/2$, by the same reasoning as the 2-seat problem. If Passenger 1 chooses seat 98, we use the same reasoning as the 3-seat problem and again get $1/2$. If Passenger 1 chooses seat 97, we apply the 4-seat solution and get $1/2$. 

What if Passenger 1 chooses seat 96 instead? This is a five-seat problem which we haven't explicitly solved. However, just as all the passengers who come before the first displaced passenger don't matter, each time a displaced passenger takes a seat that was assigned to someone else, none of the actions of the intervening passengers between *them* matter either. This fact allows us to recursively solve the problem using the simple cases we solved earlier:

\begin{align}
P(100 \rightarrow 100 | 1 \rightarrow 96) &= 1 - P(\text{96, 97, 98 or 99 $\rightarrow$ 100}) \\
&= 1 - \Bigg[P(96 \rightarrow 100) + \underbrace{P(\text{97, 98, or 99 $\rightarrow$ 100}|96\rightarrow 97)}_{\text{4-seat problem}}\cdot P(96 \rightarrow 97) \\ 
&+ \underbrace{P(\text{98 or 99 $\rightarrow$ 100}|96\rightarrow 98)}_{\text{3-seat problem}}\cdot P(96 \rightarrow 98) \\ 
&+ \underbrace{P(\text{99 $\rightarrow$ 100}|96\rightarrow 99)}_{\text{2-seat problem}}\cdot P(96 \rightarrow 99)\Bigg] \\
&= 1 - \left[\frac{1}{5} + \frac{1}{2}\cdot \frac{1}{5} + \frac{1}{2}\cdot \frac{1}{5} + \frac{1}{2}\cdot \frac{1}{5}  \right] \\
&= \boxed{\frac{1}{2}}
\end{align}

Evidently, we could solve each case recursively and we would always get an answer of $1/2$. Putting it all together to solve the 100-seat problem:

\begin{align}
P(100 \rightarrow 100) &= 1 - \Bigg[\underbrace{P(100 \rightarrow 100 |1 \rightarrow 1) \cdot P(1 \rightarrow 1)}_{=1 \cdot 1/100} + P(100 \rightarrow 100 |1 \rightarrow 2)\cdot P(1 \rightarrow 2) \\
&+ \ldots + P(100 \rightarrow 100 |1 \rightarrow 99) \cdot P(1 \rightarrow 99) + \underbrace{P(100 \rightarrow 100 |1 \rightarrow 100) \cdot P(1 \rightarrow 100)}_{= 0 \cdot 1/100}\Bigg] \\
&= 1 - \left[\frac{1}{100} + 98\cdot \frac{1}{100}\cdot \frac{1}{2} + 0 \right] \\
&= 1 - \left[\frac{100}{200}\right] \\
&= \boxed{\frac{1}{2}}
\end{align}

The probability does not depend on the total number of seats on the plane. All that matters is the fact that you are the last person in line. In other words, the probability that you get your assigned seat has nothing to do with how many people are in front of you, it only depends on how many people are *behind* you. 

You can use the same basic technique to solve the problem for the second-to-last passenger, the third-to-last passenger, etc., and the probability likewise depends only on how many people are behind them. In particular, for the $k$th-to-last passenger, the probability of them getting to sit in their assigned seat is $k/(k+1)$

## Solving by Monte Carlo Simulation

This problem lends itself very well to an object-oriented simulation. We'll define a `Passenger` object that has attributes representing the passenger's actual and assigned seat, and a method that directs them to choose a seat when it's their turn to board. Then we'll define a function that seats the passengers.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import random
import matplotlib.pyplot as plt

class Passenger:
    def __init__(self, assigned, actual=None):
        self.assigned = assigned
        self.actual = actual
            
    def choose_seat(self, available_seats):
        if self.assigned in available_seats:
            self.actual = self.assigned
        else:
            self.actual = random.choice(tuple(available_seats))
            
    def has_correct_seat(self):
        return self.assigned == self.actual
    
def load_plane(N_passengers):
    # Given a number N_passengers denoting the number
    # of passengers in line to board (= the number of seats
    # on the plane), this function runs the boarding simulation 
    # and returns a Boolean vector, with one element per 
    # passenger in line. Each element indicates whether the
    # corresponding passenger ended up with their assigned seat.
    
    # Give each passenger an assigned seat number at random
    seats = set(np.arange(1,N_passengers+1))
    passengers = [Passenger(assigned=x) for x in seats]
    random.shuffle(passengers)
    
    # Make the first passenger choose a random seat
    passengers[0].actual = random.choice(tuple(seats))
    remaining = seats - set([passengers[0].actual])
    
    # Loop through the remaining passnegers
    for p in passengers[1:]:
        p.choose_seat(remaining)
        remaining = remaining - set([p.actual])
    return np.array([p.has_correct_seat() for p in passengers])
```

```python
# Run the simulation 20,000 times for a 100-passenger plane.
N = 100
N_sim = 20000
planes = [load_plane(N) for i in np.arange(N_sim)]

# Calculate the proportion of times each passenger got their seat.
probs = np.sum(planes, axis=0)/N_sim
print(probs)
```

```python
# Calculate the predicted probabilities of each
# passenger getting their assigned seat.

# Create a vector of numbers representing each passenger's
# position in line, starting from passenger 2.
passenger_numbers = np.arange(N-1)+2

# k represents a passenger's number, counted from the back of the line.
ks = [N - x + 1 for x in passenger_numbers]
predicted_probs = [k/(k+1) for k in ks]
```

```python
# Plot the simulated and predicted results.

xvals = np.arange(1, N+1)
plt.figure(figsize = (7,5))
plt.scatter(xvals, probs, color='navy', alpha=.5, label='Simulated')
plt.plot(passenger_numbers, predicted_probs, color='black', label='Predicted')
plt.xlabel("Passenger Number")
plt.ylabel("Probability of Sitting in Assigned Seat")
plt.legend()
plt.tight_layout()
```

Our simulation yields the expected results!
